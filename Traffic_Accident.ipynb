```python
ve = dat[(dat['Vehicle']!='ไม่ระบุพาหนะ')]
```

```python
class_ve = ve['Vehicle'].value_counts()
class_ve = pd.DataFrame(class_ve)
class_ve['ชนิดของพาหนะ'] = class_ve.index
class_ve.columns = ['จำนวน','ชนิดของพาหนะ']
class_ve
```

# *Data Vizualization จำนวนการเกิดอุบัติเหตุของพาหนะแต่ละประเภท*
